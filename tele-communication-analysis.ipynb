```python
# how many person churn out(not using services)
figure=plt.figure(figsize=(3,4))
a=sns.countplot(x="Churn",data=df)
title=plt.title("Count of Customer By Churn")

a.bar_label(a.containers[0])
plt.show()
```

```python
# to show in percentage:
plt.figure(figsize=(3,4))

gb=df.groupby("Churn").agg({"Churn":"count"})
title= plt.title("Percentage Of Customer By Churn",fontsize=10)
plt.pie(gb["Churn"],autopct=("%1.2f%%"))

plt.show()
gb
```

via above chart we can conclude that 26.54% customer have been churned out... now explore the reason behind it.

```python
figure= plt.figure(figsize=(3,3))
ax=sns.countplot(x="gender",data=df,hue="Churn")
plt.title("Churn By Gender")
plt.plot(label=("Churn"))
a.bar_label(a.containers[0])
plt.show()
```

```python
plt.figure(figsize=(4,5))
a=sns.countplot(x="SeniorCitizen",data=df)
a.bar_label(a.containers[0],fontsize=7)
plt.title("Churn by Senior Citizen")
plt.show()
```

```python


# Calculate the count of churned vs non-churned for each SeniorCitizen group
df_counts = df.groupby(['SeniorCitizen', 'Churn']).size().unstack()

# Normalize the values to percentages
df_percentage = df_counts.div(df_counts.sum(axis=1), axis=0) * 100

# Plot stacked bar chart
fig, ax = plt.subplots(figsize=(4, 5))

df_percentage.plot(kind='bar', stacked=True, ax=ax)

# Adding labels on top of the bars
for p in ax.patches:
    height = p.get_height()
    if height > 0:
        ax.text(p.get_x() + p.get_width()/2 , p.get_y() + height/2 , f'{height:.1f}%', 
               ha='center', va='center',  fontsize=8, color='black')

# Title and labels
plt.title("Churn by Senior Citizen (Percentage)")
plt.ylabel("Percentage")
plt.xlabel("Senior Citizen")

plt.show()
```

#comparitively a Greater % of Senior Citizen have Churn out.

```python
fig, axes = plt.subplots(2, 4, figsize=(15, 12))

# List of the columns
columns = ['MultipleLines', 'InternetService', 'OnlineSecurity', 
           'OnlineBackup', 'DeviceProtection', 'TechSupport', 
           'StreamingTV', 'StreamingMovies']

# Flatten the axes array to easily index
axes = axes.flatten()

# Loop through each column and create a countplot
for i, column in enumerate(columns):#[enumerate funnction used to iterate list]
    sns.countplot(x=column, data=df, ax=axes[i],hue=df['Churn'])
    axes[i].set_title(f'Countplot of {column}')
    axes[i].set_xlabel(column)
    axes[i].set_ylabel('Count')

# Adjust the layout to prevent overlapping
plt.tight_layout()
plt.show()
```

people who have online security,online backup, Device Protection,Tech Support,Streaming movies and Streamimg TV having less churned out in comparision to who have multiple lines and internet service .From above plots we can conclude that person who are using Fibre Optic services are more churning out.
